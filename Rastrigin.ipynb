## Exps with Rastrigin and shifted Rastrigin

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
from PIL import Image
from plot_failedExp import plot_Exploration, plot_L_ref_fitness, plot_alpha_fitness, plot_all_alpha_dist, plot_SE_alpha_dist
import pickle
from optimizer import run
```

```python
# Select optimizers
# "SSA","PSO","GA","BAT","FFA","GWO","WOA","MVO","MFO","CS","HHO","SCA","JAYA","DE"
optimizer_list= ["GWO"] #, "DE", "PSO", "SCA"] # ["GA", "PSO", "GWO", "MVO", "WOA"] # DE, GWO, PSO, SCA, FI
```

```python
# Select benchmark function"
# "F1","F2","F3","F4","F5","F6","F7","F8","F9","F10","F11","F12","F13","F14","F15","F16","F17","F18","F19"
objectivefunc=["F9", "F24"]# Rastrigin and shifted rastrigin
```

```python
# Select number of repetitions for each experiment.
# To obtain meaningful statistical results, usually 30 independent runs are executed for each algorithm.
NumOfRuns=1 # 10 repeats of same exp
dimension= 3
```

```python
# Select general parameters for all optimizers (population size, number of iterations) ....
params = {'PopulationSize' : 50, 'Iterations' : 100} #50, 100 # 30 popu, 50 iters
```

```python
#Choose whether to Export the results in different formats
export_flags = {'Export_avg':True, 
                'Export_details':True,
                'Export_convergence':True, 
                'Export_boxplot':True}
```

```python
print(optimizer_list)
#print(len(data), len(optimizer_list))
```

```python
# Run EvoCluster # 10 dim (2 mins), 30 dim (6 mins)
out_dir= f"results_d{dimension}_r{NumOfRuns}_p{params['PopulationSize']}_i{params['Iterations']}"
final_output= run(optimizer_list, objectivefunc, NumOfRuns, params, export_flags, dimension, out_dir)

# GWO for F9 *3; GWO F24 *3; DE F9*3; DE F24*3 ...PSO, SCA
```

```python
# Save final_output to the pickle file
with open(os.path.join(out_dir, 'final_output.pkl'), 'wb') as f:
    pickle.dump(final_output, f)
```

# Plot FE Results for Single Run

```python
# ---------- Load final_output from the saved results for plotting

# Specify directory
dimension= 30 # 10, 30
NumOfRuns=1 
params = {'PopulationSize' : 50, 'Iterations' : 100}
out_dir= f"results_d{dimension}_r{NumOfRuns}_p{params['PopulationSize']}_i{params['Iterations']}"

# Refresh the varibles and reload results from pickle file
if 'final_output_loaded' in locals():
    del final_output_loaded
if 'final_output' in locals():
    del final_output
with open(os.path.join(out_dir, 'final_output.pkl'), 'rb') as f:
    final_output_loaded = pickle.load(f)

# ----------- Plot for each optimizer and objective function
for k, v in final_output_loaded.items():
    print("Optimizer and Objective Function: ",k)

    #Unpack all values
    (exp_dict, alpha_SE)= v
    print("Relative fitness for SE, FE, DE and SR: ", exp_dict)

    (alpha_SE_count, (alpha_dist_better, alpha_dist_worse), SE_alpha_dist, alpha_fitness, L_ref_fitness) = alpha_SE
    print("Number of SE with alpha: ", alpha_SE_count)
    #print("Dist moved by alpha to better fitness: ", len(alpha_dist_better), alpha_dist_better)
    #print("Dist moved by alpha to worse fitness: ", len(alpha_dist_worse), alpha_dist_worse)
    #print("Dist moved by alpha only for SE: ", len(SE_alpha_dist), SE_alpha_dist)
    #print("Fitness of alpha overtime: ", len(alpha_fitness),  alpha_fitness)
    print("Fitness of Local basin of reference (i.e. Alpha) solution overtime: ", L_ref_fitness)

    #print("exp_dict SE: ", len(exp_dict["SE"]), exp_dict["SE"])
    #print("exp_dict DE: ", len(exp_dict["DE"]), exp_dict["DE"])
    #print("exp_dict FE: ", len(exp_dict["FE"]), exp_dict["FE"])
    #print("exp_dict SR: ", len(exp_dict["SR"]), exp_dict["SR"])

    # Check num of non nan FE and SE
    print("Number of FE: ", sum(1 for i in exp_dict["FE"] if i!=None))
    print("Number of SE: ", sum(1 for i in exp_dict["SE"] if i!=None))

    plot_Exploration(exp_dict, params["Iterations"]*params["PopulationSize"], dimension)
    plot_L_ref_fitness(alpha_SE)
    plot_alpha_fitness(alpha_SE)
    plot_all_alpha_dist(alpha_SE)
    plot_SE_alpha_dist(alpha_SE)
    print("Final fitness of alpha",  alpha_fitness[(params["Iterations"]* params["PopulationSize"])-1]) # final fitness of alpha
    print("\n")
```
